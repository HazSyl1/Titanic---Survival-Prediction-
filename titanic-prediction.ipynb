```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import normalize, StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
```

```python
data_1=data.drop(['Cabin','Name'],axis='columns')
```

```python
data_1['Age'].isna().sum()
```

```python
imp = SimpleImputer(strategy='mean')
data_1['Age'] = imp.fit_transform(data_1['Age'].values.reshape(-1, 1) )

data_1['Age'].isna().sum()
```

```python
df=df.drop('Ticket',axis='columns')
```

```python
df=df.drop('PassengerId',axis='columns')
```

```python
df
```

```python
df['Sex'] = df['Sex'].astype('category')
df['Sex'] = df['Sex'].cat.codes
```

```python
df['Embarked'] = df['Embarked'].astype('category')
df['Embarked'] = df['Embarked'].cat.codes
```

```python
df.Embarked.head(5)
```

```python
df.head(5)
```

```python
df.isna().sum()
```

```python
X=df.values[:,1:9]
```

```python
Y=df.values[:,0]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=40)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train.shape[0]+X_test.shape[0] == df.shape[0]
```

```python
import time 
t=time.time()
clf_entropy=DecisionTreeClassifier(criterion='entropy',random_state=40,,)
t=time.time()-t
print(f"Time required :{t}")
```

```python
clf_entropy.fit(X_train,y_train)
```

```python
pred=clf_entropy.predict(X_test)
pred[:10]
```

```python
acc=accuracy_score(y_test,pred)
print(f"Accuracy:{acc}")
```
